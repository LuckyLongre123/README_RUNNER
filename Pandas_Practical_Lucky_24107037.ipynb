## Practical 1: Introduction to Pandas Series

### Objective:
To create and display Pandas Series using different methods.

### Concept:
A **Series** is a one-dimensional labeled array capable of holding any data type (integers, strings, floats, etc.). It is similar to a column in an Excel spreadsheet or a single column of a DataFrame.

**Key features:**
- Has an index (labels for each element)
- Can be created from lists, dictionaries, or NumPy arrays
- Supports various operations like filtering, mathematical operations, etc.

```python

import pandas as pd
import numpy as np

# Creating a Series from a list
series1 = pd.Series([10, 20, 30, 40, 50])
print("Series from List:")
print(series1)

# Creating a Series with custom index
series2 = pd.Series([100, 200, 300], index=['A', 'B', 'C'])
print("\nSeries with Custom Index:")
print(series2)

# Creating a Series from a dictionary
data_dict = {'Apple': 50, 'Banana': 30, 'Orange': 45}
series3 = pd.Series(data_dict)
print("\nSeries from Dictionary:")
print(series3)

# Accessing elements
print("\nAccessing element at index 'B':", series2['B'])
print("Accessing element at position 2:", series1[2])
```

### Explanation:

- **Series from List:** Created using `pd.Series()` with default numeric indices (0, 1, 2...).
- **Custom Index:** We can assign meaningful labels like 'A', 'B', 'C' instead of numbers.
- **Dictionary to Series:** Keys become the index, values become the data.
- **Accessing Elements:** Use index labels or numeric positions to retrieve specific values.

**Use Case:** Series are useful for representing single columns of data, like student marks in one subject or monthly sales figures.

## Practical 2: Creating and Exploring DataFrames

### Objective:
To create Pandas DataFrames and explore their basic attributes.

### Concept:
A **DataFrame** is a two-dimensional labeled data structure with columns of potentially different types. It is similar to a table in SQL or an Excel spreadsheet.

**Key attributes:**
- `shape` - Returns dimensions (rows, columns)
- `columns` - Returns column names
- `index` - Returns row labels
- `dtypes` - Returns data types of each column
- `info()` - Provides summary information
- `head()` and `tail()` - Display first/last rows

```python

import pandas as pd

# Creating a DataFrame from a dictionary
data = {
    'Name': ['Lucky', 'Karan', 'Priya', 'Rahul', 'Sneha'],
    'Age': [20, 21, 19, 22, 20],
    'Marks': [85, 90, 78, 92, 88],
    'City': ['Delhi', 'Mumbai', 'Bangalore', 'Delhi', 'Chennai']
}

df = pd.DataFrame(data)
print("DataFrame:")
print(df)

# Exploring attributes
print("\n--- DataFrame Attributes ---")
print("\nShape (rows, columns):", df.shape)
print("\nColumn Names:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nFirst 3 rows:")
print(df.head(3))
print("\nLast 2 rows:")
print(df.tail(2))

# Summary information
print("\n--- DataFrame Info ---")
df.info()
```

### Explanation:

- **DataFrame Creation:** Dictionary keys become column names, values become column data.
- **shape:** Returns (5, 4) → 5 rows and 4 columns.
- **columns:** Lists all column names.
- **dtypes:** Shows data type of each column (object for strings, int64 for integers).
- **head(n):** Displays first n rows (default 5).
- **tail(n):** Displays last n rows.
- **info():** Provides column names, non-null counts, and memory usage.

**Use Case:** DataFrames are the core structure for storing and analyzing tabular data in Pandas.

## Practical 4: Data Selection and Filtering

### Objective:
To select specific columns, rows, and filter data based on conditions.

### Concept:
Pandas provides multiple ways to select and filter data:
- **Column selection:** `df['column_name']` or `df[['col1', 'col2']]`
- **Row selection:** `df.loc[]` (label-based) or `df.iloc[]` (position-based)
- **Conditional filtering:** Boolean indexing to filter rows based on conditions

```python

import pandas as pd

# Creating a sample DataFrame
data = {
    'Name': ['Lucky', 'Karan', 'Priya', 'Rahul', 'Sneha', 'Amit', 'Neha'],
    'Age': [20, 21, 19, 22, 20, 23, 18],
    'Marks': [85, 90, 78, 92, 88, 75, 95],
    'City': ['Delhi', 'Mumbai', 'Bangalore', 'Delhi', 'Chennai', 'Mumbai', 'Delhi']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Selecting a single column
print("\n--- Name Column ---")
print(df['Name'])

# Selecting multiple columns
print("\n--- Name and Marks Columns ---")
print(df[['Name', 'Marks']])

# Selecting rows by position (iloc)
print("\n--- First 3 rows using iloc ---")
print(df.iloc[0:3])

# Selecting rows by label (loc)
print("\n--- Rows 1 to 3 using loc ---")
print(df.loc[1:3])

# Filtering: Students with marks > 85
print("\n--- Students with Marks > 85 ---")
high_scorers = df[df['Marks'] > 85]
print(high_scorers)

# Multiple conditions: Age >= 20 AND Marks > 80
print("\n--- Students: Age >= 20 AND Marks > 80 ---")
filtered = df[(df['Age'] >= 20) & (df['Marks'] > 80)]
print(filtered)

# Filtering by City
print("\n--- Students from Delhi ---")
delhi_students = df[df['City'] == 'Delhi']
print(delhi_students)
```

### Explanation:

- **Single Column:** Returns a Series.
- **Multiple Columns:** Use double brackets `[[]]` to get a DataFrame.
- **iloc:** Position-based indexing (0, 1, 2...).
- **loc:** Label-based indexing (actual row index values).
- **Boolean Filtering:** Create conditions that return True/False, then use them to filter rows.
- **Multiple Conditions:** Use `&` (AND), `|` (OR) operators. Always use parentheses around each condition.

**Use Case:** Filtering is crucial for data analysis - finding top performers, specific categories, date ranges, etc.

## Practical 5: Handling Missing Data

### Objective:
To detect, handle, and fill missing values in DataFrames.

### Concept:
Real-world data often contains missing values (NaN - Not a Number). Pandas provides several methods:
- **isnull()** / **isna():** Detect missing values
- **dropna():** Remove rows/columns with missing values
- **fillna():** Fill missing values with specific values
- **ffill()** / **bfill():** Forward fill or backward fill

```python

import pandas as pd
import numpy as np

# Creating DataFrame with missing values
data = {
    'Name': ['Lucky', 'Karan', 'Priya', None, 'Sneha', 'Amit'],
    'Age': [20, 21, np.nan, 22, 20, np.nan],
    'Marks': [85, np.nan, 78, 92, 88, 75],
    'City': ['Delhi', 'Mumbai', None, 'Delhi', 'Chennai', 'Mumbai']
}

df = pd.DataFrame(data)
print("DataFrame with Missing Values:")
print(df)

# Detecting missing values
print("\n--- Missing Values (True = Missing) ---")
print(df.isnull())

# Count of missing values per column
print("\n--- Count of Missing Values ---")
print(df.isnull().sum())

# Dropping rows with any missing value
print("\n--- After Dropping Rows with Missing Values ---")
df_dropped = df.dropna()
print(df_dropped)

# Filling missing values
print("\n--- Filling Missing Values ---")
df_filled = df.copy()
df_filled['Age'].fillna(df['Age'].mean(), inplace=True)  # Fill with mean
df_filled['Marks'].fillna(0, inplace=True)  # Fill with 0
df_filled['Name'].fillna('Unknown', inplace=True)  # Fill with 'Unknown'
df_filled['City'].fillna('Not Specified', inplace=True)
print(df_filled)

# Forward fill
print("\n--- Forward Fill (ffill) ---")
df_ffill = df.copy()
df_ffill.fillna(method='ffill', inplace=True)
print(df_ffill)
```

### Explanation:

- **isnull():** Returns True where values are missing (NaN or None).
- **isnull().sum():** Counts missing values in each column.
- **dropna():** Removes any row containing at least one missing value.
- **fillna(value):** Replaces NaN with specified value.
- **fillna(mean()):** Common strategy - fill with column average.
- **ffill (forward fill):** Fills NaN with the previous valid value.
- **bfill (backward fill):** Fills NaN with the next valid value.

**Use Case:** Cleaning real-world datasets often involves handling missing data before analysis.

## Practical 6: Data Aggregation and Grouping

### Objective:
To group data and perform aggregate operations like sum, mean, count, etc.

### Concept:
**GroupBy** is one of the most powerful features in Pandas. It allows you to:
1. **Split** data into groups based on column values
2. **Apply** functions to each group
3. **Combine** results into a new DataFrame

Common aggregate functions: `sum()`, `mean()`, `count()`, `min()`, `max()`, `std()`

```python

import pandas as pd

# Creating a sales dataset
data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Laptop', 'Mouse', 'Keyboard',
                'Laptop', 'Mouse', 'Monitor', 'Monitor'],
    'Region': ['North', 'North', 'South', 'South', 'North', 'South',
               'West', 'West', 'North', 'South'],
    'Sales': [50000, 1500, 3000, 55000, 1200, 3500, 52000, 1800, 20000, 22000],
    'Quantity': [5, 30, 20, 6, 25, 22, 4, 35, 10, 12]
}

df = pd.DataFrame(data)
print("Sales DataFrame:")
print(df)

# Group by Product and calculate total sales
print("\n--- Total Sales by Product ---")
product_sales = df.groupby('Product')['Sales'].sum()
print(product_sales)

# Group by Region and calculate average sales
print("\n--- Average Sales by Region ---")
region_avg = df.groupby('Region')['Sales'].mean()
print(region_avg)

# Multiple aggregations
print("\n--- Multiple Aggregations by Product ---")
product_agg = df.groupby('Product').agg({
    'Sales': ['sum', 'mean', 'count'],
    'Quantity': ['sum', 'mean']
})
print(product_agg)

# Group by multiple columns
print("\n--- Sales by Product and Region ---")
multi_group = df.groupby(['Product', 'Region'])['Sales'].sum()
print(multi_group)

# Count of products sold per region
print("\n--- Product Count by Region ---")
region_count = df.groupby('Region')['Product'].count()
print(region_count)
```

### Explanation:

- **groupby('column'):** Groups data by unique values in that column.
- **['Sales'].sum():** After grouping, apply sum() to the Sales column.
- **agg():** Apply multiple aggregation functions at once.
- **Multiple groupby:** Group by more than one column to create hierarchical groups.
- **count():** Counts number of items in each group.

**Use Case:** Business analytics - calculating total sales by product, average scores by class, revenue by region, etc.

## Practical 7: Sorting and Ranking Data

### Objective:
To sort DataFrames by column values and rank data.

### Concept:
Sorting helps in organizing data for better analysis and visualization:
- **sort_values():** Sort by one or more columns
- **sort_index():** Sort by row index
- **rank():** Assign ranks to values

You can sort in ascending or descending order.

```python

import pandas as pd

# Creating a student performance dataset
data = {
    'Name': ['Lucky', 'Karan', 'Priya', 'Rahul', 'Sneha', 'Amit', 'Neha'],
    'Marks': [85, 90, 78, 92, 88, 75, 95],
    'Age': [20, 21, 19, 22, 20, 23, 18],
    'City': ['Delhi', 'Mumbai', 'Bangalore', 'Delhi', 'Chennai', 'Mumbai', 'Delhi']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Sort by Marks (ascending)
print("\n--- Sorted by Marks (Ascending) ---")
sorted_asc = df.sort_values('Marks')
print(sorted_asc)

# Sort by Marks (descending)
print("\n--- Sorted by Marks (Descending) ---")
sorted_desc = df.sort_values('Marks', ascending=False)
print(sorted_desc)

# Sort by multiple columns
print("\n--- Sorted by City, then Marks ---")
multi_sort = df.sort_values(['City', 'Marks'], ascending=[True, False])
print(multi_sort)

# Adding rank column
print("\n--- DataFrame with Rank ---")
df['Rank'] = df['Marks'].rank(ascending=False, method='min')
df_sorted = df.sort_values('Rank')
print(df_sorted)

# Top 3 students
print("\n--- Top 3 Students ---")
top3 = df.nlargest(3, 'Marks')
print(top3[['Name', 'Marks']])

# Bottom 3 students
print("\n--- Bottom 3 Students ---")
bottom3 = df.nsmallest(3, 'Marks')
print(bottom3[['Name', 'Marks']])
```

### Explanation:

- **sort_values('column'):** Sorts DataFrame by specified column (default ascending).
- **ascending=False:** Sorts in descending order.
- **Sort by multiple columns:** First sorts by first column, then by second for ties.
- **rank():** Assigns ranks (1 = highest if ascending=False).
- **nlargest(n):** Returns top n rows by column value.
- **nsmallest(n):** Returns bottom n rows.

**Use Case:** Finding top performers, ranking students/products, organizing leaderboards.

## Practical 8: Merging and Joining DataFrames

### Objective:
To combine multiple DataFrames using merge, join, and concatenate operations.

### Concept:
Often data is spread across multiple tables/files. Pandas provides several methods to combine them:
- **concat():** Stack DataFrames vertically or horizontally
- **merge():** SQL-like joins (inner, outer, left, right)
- **join():** Join on index

This is similar to SQL JOIN operations.

```python

import pandas as pd

# Creating two related DataFrames
students = pd.DataFrame({
    'Student_ID': [101, 102, 103, 104],
    'Name': ['Lucky', 'Karan', 'Priya', 'Rahul'],
    'Age': [20, 21, 19, 22]
})

marks = pd.DataFrame({
    'Student_ID': [101, 102, 103, 105],
    'Subject': ['Math', 'Science', 'Math', 'English'],
    'Marks': [85, 90, 78, 88]
})

print("Students DataFrame:")
print(students)
print("\nMarks DataFrame:")
print(marks)

# Inner Join (only matching records)
print("\n--- Inner Join ---")
inner_merged = pd.merge(students, marks, on='Student_ID', how='inner')
print(inner_merged)

# Left Join (all from left, matching from right)
print("\n--- Left Join ---")
left_merged = pd.merge(students, marks, on='Student_ID', how='left')
print(left_merged)

# Right Join
print("\n--- Right Join ---")
right_merged = pd.merge(students, marks, on='Student_ID', how='right')
print(right_merged)

# Outer Join (all records from both)
print("\n--- Outer Join ---")
outer_merged = pd.merge(students, marks, on='Student_ID', how='outer')
print(outer_merged)

# Concatenating DataFrames vertically
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
print("\n--- Concatenation (Vertical) ---")
concat_result = pd.concat([df1, df2], ignore_index=True)
print(concat_result)
```

### Explanation:

- **Inner Join:** Returns only rows where Student_ID exists in both DataFrames.
- **Left Join:** Returns all rows from left DataFrame, fills NaN where right doesn't match.
- **Right Join:** Returns all rows from right DataFrame.
- **Outer Join:** Returns all rows from both, fills NaN where no match.
- **concat():** Stacks DataFrames vertically (axis=0) or horizontally (axis=1).

**Use Case:** Combining data from multiple sources - student info from one file, marks from another; sales data from different months; customer data with transaction history.

## Practical 9: Data Visualization with Pandas

### Objective:
To create basic visualizations directly from Pandas DataFrames.

### Concept:
Pandas has built-in plotting capabilities using Matplotlib. You can create:
- Line plots
- Bar charts
- Histograms
- Scatter plots
- Box plots

This provides quick visual insights without writing complex Matplotlib code.

```python

import pandas as pd
import matplotlib.pyplot as plt

# Creating a sample dataset
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [45000, 50000, 48000, 62000, 58000, 70000],
    'Expenses': [30000, 32000, 31000, 40000, 38000, 45000]
}

df = pd.DataFrame(data)
print("Sales Data:")
print(df)

# Line plot
print("\n--- Creating Line Plot ---")
df.plot(x='Month', y=['Sales', 'Expenses'], kind='line',
        marker='o', title='Monthly Sales vs Expenses', figsize=(10, 5))
plt.ylabel('Amount (₹)')
plt.grid(True)
plt.show()

# Bar chart
print("\n--- Creating Bar Chart ---")
df.plot(x='Month', y='Sales', kind='bar', color='skyblue',
        title='Monthly Sales', figsize=(10, 5))
plt.ylabel('Sales (₹)')
plt.xticks(rotation=0)
plt.show()

# Creating student marks data for more plots
marks_data = {
    'Student': ['Student' + str(i) for i in range(1, 21)],
    'Math': [85, 78, 90, 67, 88, 92, 74, 81, 95, 70,
             86, 79, 91, 68, 89, 93, 75, 82, 94, 71],
    'Science': [80, 85, 88, 72, 86, 90, 78, 83, 92, 75,
                81, 84, 89, 73, 87, 91, 79, 82, 93, 76]
}

df_marks = pd.DataFrame(marks_data)

# Histogram
print("\n--- Creating Histogram ---")
df_marks[['Math', 'Science']].plot(kind='hist', bins=10, alpha=0.7,
                                    title='Distribution of Marks', figsize=(10, 5))
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()

# Box plot
print("\n--- Creating Box Plot ---")
df_marks[['Math', 'Science']].plot(kind='box',
                                    title='Marks Distribution', figsize=(10, 5))
plt.ylabel('Marks')
plt.show()

# Scatter plot
print("\n--- Creating Scatter Plot ---")
df_marks.plot(kind='scatter', x='Math', y='Science',
              title='Math vs Science Marks', figsize=(8, 6),
              color='green', alpha=0.6, s=100)
plt.grid(True)
plt.show()
```

### Explanation:

- **Line Plot:** Shows trends over time (months). Useful for time series data.
- **Bar Chart:** Compares values across categories. Good for categorical data.
- **Histogram:** Shows distribution/frequency of continuous data (marks range).
- **Box Plot:** Displays quartiles, median, and outliers. Useful for understanding data spread.
- **Scatter Plot:** Shows relationship between two numerical variables.

**Parameters:**
- `kind`: Type of plot ('line', 'bar', 'hist', 'box', 'scatter')
- `figsize`: Size of figure (width, height) in inches
- `alpha`: Transparency (0 to 1)
- `color`: Color of plot elements

**Use Case:** Quick visual analysis of data patterns, trends, distributions, and relationships without writing extensive plotting code.

## Practical 10: Real-World Application - Student Performance Analysis

### Objective:
To perform comprehensive analysis of student performance data using all Pandas concepts learned.

### Concept:
This practical combines:
- Data creation and loading
- Data cleaning (handling missing values)
- Filtering and selection
- Grouping and aggregation
- Sorting and ranking
- Statistical analysis
- Visualization

This demonstrates how Pandas is used in real-world data analysis scenarios.

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating comprehensive student dataset
np.random.seed(42)

students_data = {
    'Student_ID': range(1, 51),
    'Name': [f'Student_{i}' for i in range(1, 51)],
    'Class': np.random.choice(['10A', '10B', '10C'], 50),
    'Math': np.random.randint(40, 100, 50),
    'Science': np.random.randint(40, 100, 50),
    'English': np.random.randint(40, 100, 50),
    'Social_Studies': np.random.randint(40, 100, 50),
    'Computer': np.random.randint(40, 100, 50),
    'Attendance_%': np.random.randint(60, 100, 50)
}

df = pd.DataFrame(students_data)

# Introducing some missing values randomly
df.loc[np.random.choice(df.index, 5), 'Math'] = np.nan
df.loc[np.random.choice(df.index, 3), 'Science'] = np.nan

print("="*60)
print("  STUDENT PERFORMANCE ANALYSIS SYSTEM")
print("="*60)
print("\n📊 Dataset Overview:")
print(df.head(10))
print(f"\nTotal Students: {len(df)}")
print(f"\nDataset Shape: {df.shape}")

# Data Info
print("\n--- Dataset Information ---")
df.info()
```

```python
# STEP 1: Data Cleaning - Handling Missing Values
print("\n" + "="*60)
print("STEP 1: DATA CLEANING")
print("="*60)

print("\n Missing Values Count:")
print(df.isnull().sum())

# Fill missing values with subject mean
df['Math'].fillna(df['Math'].mean(), inplace=True)
df['Science'].fillna(df['Science'].mean(), inplace=True)

print("\n Missing values filled with subject averages")
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

# STEP 2: Calculate Total and Percentage
print("\n" + "="*60)
print("STEP 2: CALCULATING TOTAL MARKS AND PERCENTAGE")
print("="*60)

df['Total_Marks'] = df[['Math', 'Science', 'English', 'Social_Studies', 'Computer']].sum(axis=1)
df['Percentage'] = (df['Total_Marks'] / 500) * 100
df['Grade'] = pd.cut(df['Percentage'],
                     bins=[0, 40, 60, 75, 90, 100],
                     labels=['F', 'C', 'B', 'A', 'A+'])

print("\n Updated DataFrame with Total, Percentage, and Grade:")
print(df[['Name', 'Total_Marks', 'Percentage', 'Grade']].head(10))
```

```python

# STEP 3: Statistical Analysis
print("\n" + "="*60)
print("STEP 3: STATISTICAL ANALYSIS")
print("="*60)

print("\n Overall Statistics:")
stats = df[['Math', 'Science', 'English', 'Social_Studies', 'Computer', 'Percentage']].describe()
print(stats)

print("\n Subject-wise Average Marks:")
subject_avg = df[['Math', 'Science', 'English', 'Social_Studies', 'Computer']].mean()
print(subject_avg.sort_values(ascending=False))

print("\n Top 5 Students (by Percentage):")
top5 = df.nlargest(5, 'Percentage')[['Name', 'Class', 'Total_Marks', 'Percentage', 'Grade']]
print(top5)

print("\n Bottom 5 Students (by Percentage):")
bottom5 = df.nsmallest(5, 'Percentage')[['Name', 'Class', 'Total_Marks', 'Percentage', 'Grade']]
print(bottom5)
```

```python

# STEP 4: Class-wise Analysis
print("\n" + "="*60)
print("STEP 4: CLASS-WISE PERFORMANCE ANALYSIS")
print("="*60)

class_performance = df.groupby('Class').agg({
    'Total_Marks': ['mean', 'max', 'min'],
    'Percentage': 'mean',
    'Attendance_%': 'mean',
    'Student_ID': 'count'
}).round(2)

class_performance.columns = ['Avg_Total', 'Max_Total', 'Min_Total',
                              'Avg_Percentage', 'Avg_Attendance', 'Student_Count']
print("\n Class-wise Summary:")
print(class_performance)

print("\n Grade Distribution by Class:")
grade_dist = pd.crosstab(df['Class'], df['Grade'])
print(grade_dist)

# STEP 5: Subject-wise Analysis
print("\n" + "="*60)
print("STEP 5: SUBJECT-WISE ANALYSIS")
print("="*60)

subjects = ['Math', 'Science', 'English', 'Social_Studies', 'Computer']

print("\n Students scoring above 90 in each subject:")
for subject in subjects:
    high_scorers = df[df[subject] >= 90][['Name', subject]].shape[0]
    print(f"{subject}: {high_scorers} students")

print("\n Students scoring below 50 in each subject:")
for subject in subjects:
    low_scorers = df[df[subject] < 50][['Name', subject]].shape[0]
    print(f"{subject}: {low_scorers} students")
```

```python

# STEP 6: Filtering and Advanced Analysis
print("\n" + "="*60)
print("STEP 6: ADVANCED FILTERING")
print("="*60)

# High performers (>80% and >90% attendance)
print("\n High Performers (>80% marks, >90% attendance):")
high_performers = df[(df['Percentage'] > 80) & (df['Attendance_%'] > 90)]
print(f"Total: {len(high_performers)} students")
print(high_performers[['Name', 'Class', 'Percentage', 'Attendance_%', 'Grade']].head())

# Students needing attention (<60%)
print("\n Students Needing Attention (<60%):")
need_attention = df[df['Percentage'] < 60]
print(f"Total: {len(need_attention)} students")
print(need_attention[['Name', 'Class', 'Percentage', 'Grade']].head())

# Best class based on average percentage
print("\n Best Performing Class:")
best_class = df.groupby('Class')['Percentage'].mean().idxmax()
best_avg = df.groupby('Class')['Percentage'].mean().max()
print(f"Class: {best_class} with Average: {best_avg:.2f}%")

# Correlation between Attendance and Performance
print("\n Correlation between Attendance and Percentage:")
correlation = df['Attendance_%'].corr(df['Percentage'])
print(f"Correlation Coefficient: {correlation:.3f}")
if correlation > 0.5:
    print(" Strong positive correlation - Better attendance leads to better performance!")
else:
    print(" Weak correlation")
```

```python

# STEP 7: Visualizations
print("\n" + "="*60)
print("STEP 7: DATA VISUALIZATION")
print("="*60)

# 1. Subject-wise average marks
plt.figure(figsize=(10, 6))
subject_avg.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Subject-wise Average Marks', fontsize=16, fontweight='bold')
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Average Marks', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# 2. Grade distribution
plt.figure(figsize=(8, 6))
df['Grade'].value_counts().sort_index().plot(kind='bar', color='coral', edgecolor='black')
plt.title('Grade Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Grade', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# 3. Class-wise performance comparison
plt.figure(figsize=(10, 6))
df.groupby('Class')['Percentage'].mean().plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Class-wise Average Percentage', fontsize=16, fontweight='bold')
plt.xlabel('Class', fontsize=12)
plt.ylabel('Average Percentage', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# 4. Distribution of percentage
plt.figure(figsize=(10, 6))
plt.hist(df['Percentage'], bins=15, color='purple', alpha=0.7, edgecolor='black')
plt.title('Distribution of Student Percentages', fontsize=16, fontweight='bold')
plt.xlabel('Percentage', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# 5. Attendance vs Percentage scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Attendance_%'], df['Percentage'], alpha=0.6, c='blue', s=50)
plt.title('Attendance vs Performance', fontsize=16, fontweight='bold')
plt.xlabel('Attendance %', fontsize=12)
plt.ylabel('Percentage', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 6. Box plot for all subjects
plt.figure(figsize=(12, 6))
df[subjects].plot(kind='box', color='lightblue')
plt.title('Subject-wise Marks Distribution', fontsize=16, fontweight='bold')
plt.ylabel('Marks', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python

# STEP 8: Export Results
print("\n" + "="*60)
print("STEP 8: EXPORTING RESULTS")
print("="*60)

# Save cleaned and analyzed data
df.to_csv('student_performance_analysis.csv', index=False)
print("\n Complete analysis saved to 'student_performance_analysis.csv'")

# Save top performers
top_performers = df[df['Grade'].isin(['A', 'A+'])][['Name', 'Class', 'Total_Marks', 'Percentage', 'Grade']]
top_performers.to_csv('top_performers.csv', index=False)
print(" Top performers list saved to 'top_performers.csv'")

# Final Summary Report
print("\n" + "="*60)
print("FINAL SUMMARY REPORT")
print("="*60)

print(f"""
  OVERALL STATISTICS:
   • Total Students Analyzed: {len(df)}
   • Average Percentage: {df['Percentage'].mean():.2f}%
   • Highest Percentage: {df['Percentage'].max():.2f}%
   • Lowest Percentage: {df['Percentage'].min():.2f}%
   • Average Attendance: {df['Attendance_%'].mean():.2f}%

  GRADE DISTRIBUTION:
{df['Grade'].value_counts().to_string()}

  CLASS PERFORMANCE:
{df.groupby('Class')['Percentage'].mean().round(2).to_string()}

  KEY INSIGHTS:
   • Best Performing Subject: {subject_avg.idxmax()} ({subject_avg.max():.2f})
   • Weakest Subject: {subject_avg.idxmin()} ({subject_avg.min():.2f})
   • Attendance-Performance Correlation: {correlation:.3f}
   • Students with A/A+ Grade: {len(df[df['Grade'].isin(['A', 'A+'])])}
   • Students Needing Attention: {len(need_attention)}
""")

print("="*60)
print("  ANALYSIS COMPLETE ")
print("="*60)
```
